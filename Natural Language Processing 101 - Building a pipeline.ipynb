#### Step 4: Text Lemmatization/Stemming

- Stemming is the process of reducing inflection in words to their root forms such as mapping a group of words to the same stem even if the stem itself is not a valid word in the Language. Stem (root) is the part of the word to which you add inflectional (changing/deriving) affixes such as (-ed,-ize, -s,-de,mis).


- Lemmatization, unlike Stemming, reduces the inflected words properly ensuring that the root word belongs to the language. In Lemmatization root word is called Lemma. A lemma (plural lemmas or lemmata) is the canonical form, dictionary form, or citation form of a set of words.


```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

##

sentence = 'runs running ran , plays playing played , am are is'
print(sentence)

sentence = sentence.split()


porter = PorterStemmer()
stem_sentence =[]
for word in sentence:
    stem_sentence.append(porter.stem(word))
    stem_sentence.append(" ")
    
print("Stemmed:",''.join(stem_sentence))

wordnet = WordNetLemmatizer()
lemma_sentence =[]
for word in sentence:
    lemma_sentence.append(wordnet.lemmatize(word, pos="v"))
    lemma_sentence.append(" ")
    
print("Lemmatized:",''.join(lemma_sentence))
```

```python
print(porter.stem('was'), wordnet.lemmatize('was', pos="v"))

print(porter.stem('went'), wordnet.lemmatize('went', pos="v"))
```

```python
lemmatized_sentence =[]
for sentence in tokenized_sentences:
    temp = []
    for word in sentence:
        temp.append(wordnet.lemmatize(word))
        temp.append(" ")
    
    lemmatized_sentence.append(''.join(temp))
    
```

#### Step 5: Identifying Stop Words

```python
from nltk.corpus import stopwords
#set(stopwords.words('english'))
    
```

```python
stopwords_removed = []
for sentence in lemmatized_sentence:
    sentence = sentence.split()
    temp = []
    for word in sentence:
        
        if word not in stopwords.words():
            temp.append(word)
           
    stopwords_removed.append(' '.join(temp))
            
```

#### Step 6: Dependency Parsing

The goal is to build a tree that assigns a single parent word to each word in the sentence. The root of the tree will be the main verb in the sentence.

And if we followed the complete parse tree for the sentence (beyond what is shown), we would even found out that London is the capital of the United Kingdom.

#### Step 8: Coreference Resolution

Coreference resolution is one of the most difficult steps in our pipeline to implement. It’s even more difficult than sentence parsing. Recent advances in deep learning have resulted in new approaches that are more accurate, but it isn’t perfect yet.

As a human reading this sentence, you can easily figure out that “it” means “London”. The goal of coreference resolution is to figure out this same mapping by tracking pronouns across sentences. We want to figure out all the words that are referring to the same entity.
